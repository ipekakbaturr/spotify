```python
#top 10 songs popularity > 90 
popular = tracks.query('popularity>90', inplace=False).sort_values('popularity',ascending=False)
popular.head(10)
```

```python
tracks.set_index('release_date',inplace=True)
tracks.index = pd.to_datetime(tracks.index, format= "mixed",yearfirst=True)
tracks.head()
```

```python
tracks[['artists']].iloc[18]
```

```python
#convert ms to second
tracks['duration']=tracks['duration_ms'].apply(lambda x:round(x/1000))
tracks.drop('duration_ms', inplace=True,axis=1)
```

```python
tracks.duration.head(5)
```

```python
#correlation map with pearson correlation method
corr_df= tracks.drop(['key','mode','explicit'],axis=1)
corr_df.head()
```

```python
plt.figure(figsize=(14,6))
heatmap=sns.heatmap(corr_df.corr(method='pearson',numeric_only=True),annot=True,fmt='.1g',vmin=-1,vmax=1,center=0,cmap='inferno', linewidths=1,linecolor='Black')
heatmap.set_title('Pearson Correlation Heatmap')
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation=90)
```

Total number of songs each year from 1922

```python
tracks['dates']= tracks.index.get_level_values('release_date')
tracks.dates= pd.to_datetime(tracks.dates)
years=tracks.dates.dt.year
```

```python
sns.displot(years,discrete=True,aspect=2,height=5,kind='hist').set(title=' Total Number of Songs Released Each Year')
```

Duration of songs over years

```python
total_duration=tracks.duration
fig_dims=(18,7)
fig,ax=plt.subplots(figsize=fig_dims)
fig=sns.barplot(x=years,y=total_duration, ax=ax,palette='flare').set(title='Duration of Songs')
plt.xticks(rotation=90)
```

```python
total_duration=tracks.duration
fig_dims=(10,5)
fig,ax=plt.subplots(figsize=fig_dims)
fig=sns.lineplot(x=years,y=total_duration, ax=ax,palette='flare').set(title='Duration of Songs')
plt.xticks(rotation=60)
```

Duration of songs declining since attention time of the listeners are declining!

```python
genre=pd.read_csv('Downloads/SpotifyFeatures.csv')
genre.head()
```

Duration of songs in different genres:

```python
plt.title('Duration of Songs in Different Genres')
sns.barplot(y='genre',x='duration_ms',data=genre,palette='rocket')
plt.xlabel('Duration in ms')
plt.ylabel('Genres')
```

```python
sns.set_style(style='darkgrid')
plt.figure(figsize=(10,5))
famous=genre.sort_values('popularity',ascending=False).head(10)
sns.barplot(y='genre',x='popularity',data=famous,palette='rainbow').set(title='Top 5 Genres')
```
